```python
import pandas as pd
import os

# Define the file path
file_path = r"QQQ_2025-04-17_EOD_Calls_Puts_2024-05-13_to_2025-04-11.csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display basic information about the DataFrame
print(f"DataFrame shape: {df.shape}")
print("\nFirst 5 rows:")
display(df.head())

# Check for missing values
print("\nMissing values per column:")
display(df.isna().sum())

# Display data types
print("\nData types:")
display(df.dtypes)
```

```python
# Check for duplicate rows
print(f"Shape before removing duplicates: {df.shape}")

# Drop duplicate rows
df_cleaned = df.drop_duplicates()

# Display shape after removing duplicates
print(f"Shape after removing duplicates: {df_cleaned.shape}")

# Calculate how many duplicates were removed
duplicates_removed = df.shape[0] - df_cleaned.shape[0]
print(f"Number of duplicate rows removed: {duplicates_removed}")

# Update our main dataframe
df = df_cleaned
```

```python
# Define columns to drop
columns_to_drop = [
    'contract',
    'underlying_symbol',
    'exp_date',
    'expiration_type',
    'exchange',
    'currency',
    'bid_date',
    'ask_date',
    'tradetime',
    'record_id',
    'record_type'
]

# Drop the specified columns
print(f"Shape before dropping columns: {df.shape}")
df = df.drop(columns=columns_to_drop, errors='ignore')
print(f"Shape after dropping columns: {df.shape}")

# Display the first few rows to verify columns were dropped
print("\nDataFrame after dropping columns:")
display(df.head())
```

```python
# Sort the DataFrame by trade_date in descending order
df = df.sort_values(by='trade_date', ascending=False)

# Display the first few rows to verify sorting
print("\nDataFrame after sorting by trade_date (descending):")
display(df.head())

# Check the unique trade dates to confirm sorting worked
print("\nUnique trade dates (first 10):")
print(df['trade_date'].unique()[:10])
```

```python
# Create binary columns for call/put as integer type
print("Converting 'type' column to binary integers...")
df['call'] = (df['type'] == 'call').astype(int)
df['put'] = (df['type'] == 'put').astype(int)

# Drop the original 'type' column
df = df.drop(columns=['type'])

# Display the first few rows to verify the conversion
print("\nDataFrame after converting 'type' to binary integers:")
display(df.head())

# Check the shape to confirm the transformation
print(f"\nShape after type conversion: {df.shape}")
```

```python
# Calculate the percentage of rows with volume less than 10
low_volume_count = (df['volume'] < 10).sum()
total_rows = len(df)
low_volume_percentage = (low_volume_count / total_rows) * 100

# Display the results
print(f"\nAnalyzing volume data:")
print(f"Total number of rows: {total_rows}")
print(f"Number of rows with volume < 10: {low_volume_count}")
print(f"Percentage of rows with volume < 10: {low_volume_percentage:.2f}%")

# Optional: Display a sample of low volume rows
print("\nSample of rows with volume < 10:")
display(df[df['volume'] < 10].head())
```

```python
# Filter out rows with volume <= 9
print("\nRemoving rows with volume <= 9...")
original_row_count = len(df)
df = df[df['volume'] > 9]
new_row_count = len(df)
removed_rows = original_row_count - new_row_count

# Display the results
print(f"Original number of rows: {original_row_count}")
print(f"Number of rows after filtering: {new_row_count}")
print(f"Number of rows removed: {removed_rows}")
print(f"Percentage of rows removed: {(removed_rows / original_row_count) * 100:.2f}%")

# Display the first few rows of the filtered DataFrame
print("\nDataFrame after removing low volume rows:")
display(df.head())

# Verify that no rows with volume <= 9 remain
remaining_low_volume = (df['volume'] <= 9).sum()
print(f"\nNumber of rows with volume <= 9 remaining: {remaining_low_volume}")
```

```python
# Save the cleaned DataFrame to a CSV file
output_path = 'QQQ_Chain_Clean.csv'
print(f"\nSaving cleaned data to {output_path}...")

# Save to CSV without the index column
df.to_csv(output_path, index=False)

print(f"Data successfully saved to {output_path}")
print(f"File contains {len(df)} rows and {len(df.columns)} columns")

# Optional: Display file information
import os
if os.path.exists(output_path):
    file_size_kb = os.path.getsize(output_path) / 1024
    print(f"File size: {file_size_kb:.2f} KB")
```
